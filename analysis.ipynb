```python
import pyrosm
```

```python
# Initialize the OSM object
fp = pyrosm.get_data("Dublin")
osm = pyrosm.OSM(fp)
```

```python
type(osm)
```

```python
drive_net = osm.get_network(network_type="driving")
drive_net.plot()
```

```python
walk_net = osm.get_network(network_type="walking")
walk_net.plot()
```

```python
walk_net.plot(color="k", figsize=(12,12), lw=0.7, alpha=0.6)
```

```python
# Download the PBF file if you haven't already
fp = pyrosm.get_data("italy")

# Initialize the OSM object
osm = pyrosm.OSM(fp)

# Define a custom filter
custom_filter = {"name:en": ["Padua"]}

# Read the data with the custom filter
df = osm.get_data_by_custom_criteria(custom_filter = custom_filter)
```

```python
walk_net = osm.get_network(network_type="walking")
walk_net.plot(color="k", figsize=(12,12), lw=0.7, alpha=0.6)
```
